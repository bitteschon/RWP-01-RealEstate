1. 단지명에서 번지 삭제하기
    - (숫자-숫자) 형태 : re 사용하자

```python
# re 쓰기전, nan 값이 있는지 확인
df['단지명'].isnull().values.any()
```

```python
df['단지명(clean)'] = df.단지명.str.replace(r'[(]\d+[-]\d+[)]', '').str.replace(r'(\d+[-]\d+)', '')
df
```

```python
def findApartname(x):
    if x.find('(') > 0:
        return x.split('(')[0]
    else :
        return x
    
df['단지명(clean)'] = df['단지명(clean)'].apply(findApartname)
df      
```

```python
df['단지명(clean)'].value_counts()
```

2. 칼럼명 변경
    - '군' > '구', '구' > '동'

```python
df = df.rename({'구':'동'}, axis='columns')
df.head()
```

```python
df = df.rename({'군':'구', }, axis='columns')
df.head()
```

3. 칼럼 합치기 
    - '시'+'구'+'동'+'단지명(clean)'

```python
df['uid'] = df['시'] + ' ' + df['구'] + ' ' + df['동'] + ' ' + df['단지명(clean)']
df
```
